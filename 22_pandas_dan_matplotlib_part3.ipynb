```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/boedybios/kaggle_explorations/master/Iris_Data_Classification_and_EDA/dataset/iris/Iris.csv'
df = pd.read_csv(url)

df.head()
```

```python
df[['SepalLengthCm', 'SepalWidthCm',
    'PetalLengthCm', 'PetalWidthCm']].plot(
    kind='hist',
    bins=10,
    subplots=True,
    layout=(2, 2),
    figsize=(8, 6)
)

plt.suptitle('Contoh Subplots Histogram Dataset Iris')
plt.tight_layout()
plt.show()
```
